```python
import requests
```

```python
api_dos_paises = 'https://restcountries.com/v3.1/all?fields=name,capital'
```

```python
resposta = requests.get(api_dos_paises)
```

```python
dados = resposta.json()
```

```python
print(dados)
```

```python
from rich import print_json
import json
```

```python
print_json(data=dados)
```

```python
paises = []
```

```python
for pais in dados:
  paises.append(
      {
          'nome': pais.get('name', {}).get('common'),
          'capital': pais.get('capital')
      }
  )
```

```python
print(paises)
```

```python
response = requests.get(url_paises_da_europa)
dados_europa = response.json()
```

```python
print(dados_europa)
```

```python
url_dependentes_latinos = 'https://restcountries.com/v3.1/independent?status=false&fields=name,region'
```

```python
response = requests.get(url_dependentes_latinos)
dados_latinos_dependentes = response.json()
```

```python
print(dados_latinos_dependentes)
```

```python
latinos_colonizados = []
```

```python
for pais in dados_latinos_dependentes:
  if pais.get('region') == 'Americas':
    latinos_colonizados.append(
        {
            'nome': pais.get('name', {}).get('common'),
            'regiao': pais.get('region')
        }
    )
```

```python
print(latinos_colonizados)
```
