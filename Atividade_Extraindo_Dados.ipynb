```python
import requests

# Exercício 1
def todos_paises():
    """
    Busca dados de todos os países na API restcountries.com (v2)
    Retorna uma lista de dicionários com nome, população e região
    """
    url = "https://restcountries.com/v3.1/all?fields=name,population,region"
    response = requests.get(url)
    response.raise_for_status()
    dados = response.json()

    lista = []
    for pais in dados:
        lista.append({
            "nome": pais["name"],              # v2 usa apenas "name"
            "populacao": pais.get("population", "N/A"),
            "regiao": pais.get("region", "N/A")
        })
    return lista

# Teste: mostrar os 5 primeiros
print("Exercício 1 - Lista de países (nome, população, região):")
for pais in todos_paises()[:5]:
    print(pais)
```

```python
import requests

def paises_africa():
    """
    Busca todos os países da África usando a API REST Countries.
    Retorna uma lista com os nomes comuns dos países.
    """
    url = "https://restcountries.com/v3.1/region/africa"
    response = requests.get(url)
    dados = response.json()

    return [pais["name"]["common"] for pais in dados]

# Agora chamamos a função fora dela
todos_africa = paises_africa()
print("\nTodos os países da África:")
print(todos_africa)
```

```python
def paises_africa_independentes():
    url = "https://restcountries.com/v3.1/region/africa"
    response = requests.get(url)
    dados = response.json()

    independentes = []

    for pais in dados:
        if pais.get("independent") == True:
            independentes.append(pais["name"]["common"])

    return independentes

resultado = paises_africa_independentes()
print(resultado[:10])
```

```python
import requests
import csv   # faltava esse import

def salvar_paises_csv():
    """
    Busca todos os países na API REST Countries,
    limpa e padroniza os dados e salva em CSV.
    """
    url = "https://restcountries.com/v3.1/all?fields=name,population,region"
    response = requests.get(url)
    response.raise_for_status()
    dados = response.json()

    paises = []

    for pais in dados:
        paises.append({
            "nome": pais.get("name", {}).get("common", "").upper(),   # nome em maiúsculas
            "populacao": pais.get("population", 0),                   # população como número
            "regiao": pais.get("region", "").lower()                  # região em minúsculas
        })

    # salva em CSV
    with open("paises.csv", "w", newline="", encoding="utf-8") as arquivo:
        writer = csv.DictWriter(
            arquivo,
            fieldnames=["nome", "populacao", "regiao"]
        )
        writer.writeheader()
        writer.writerows(paises)

# Executa a função
salvar_paises_csv()
print("Arquivo 'paises.csv' criado com sucesso!")
```

```python
import csv

def ler_paises_csv():
    """
    Lê o arquivo paises.csv e imprime os primeiros registros
    """
    with open("paises.csv", "r", encoding="utf-8") as arquivo:
        reader = csv.DictReader(arquivo)
        dados = list(reader)

    # Mostrar os 10 primeiros países
    print("\nPrimeiros 10 países do CSV:")
    for pais in dados[:10]:
        print(pais)

# Executa a leitura
ler_paises_csv()
```
